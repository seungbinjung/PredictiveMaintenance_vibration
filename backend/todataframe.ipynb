```python
import sys
from pathlib import Path
import os

# Jupyter 노트북에서는 __file__이 없으므로 현재 작업 디렉토리 기반으로 경로 설정
# 노트북이 backend 디렉토리에 있으므로, 상위 디렉토리(caps)를 프로젝트 루트로 설정
PROJECT_ROOT = Path(os.getcwd()).resolve().parent  # backend -> caps
EXTRA_PATHS = [
    str(PROJECT_ROOT),                    # caps 루트
    str(PROJECT_ROOT / "backend"),        # backend 디렉토리
#     str(PROJECT_ROOT / "ByteTrack"),        # ByteTrack 직접 참조가 필요한 경우
#     str(PROJECT_ROOT / "deep-person-reid-master"),  # 필요한 경우만
#     str(PROJECT_ROOT / "app" / "models" / "mapping"),   # point_transformer 경로 수정
]

for p in EXTRA_PATHS:
    if p not in sys.path:
        sys.path.insert(0, p)
        
print(f"프로젝트 루트: {PROJECT_ROOT}")
print(f"추가된 경로: {EXTRA_PATHS}")
```
